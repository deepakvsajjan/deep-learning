```python
from datasets import load_dataset
import pandas as pd
raw_data = load_dataset("wangrongsheng/ag_news")
train_data = load_dataset("wangrongsheng/ag_news", split="train")
print(train_data[0])
```

cleanig the datasets

```python
import re
def clean_text(example):
    text = example["text"].lower()
    text = re.sub(r"\d+", "", text)
    text = re.sub(r"[^a-z\s]", "", text)
    text = re.sub(r"\s+", " ", text).strip()
    example["clean_text"] = text
    return example
dataset = train_data.map(clean_text)
```

Tokenize the dataset into the number

```python
from tensorflow.keras.preprocessing.text import Tokenizer
tokenizer = Tokenizer(oov_token="<OOV>")
tokenizer.fit_on_texts(dataset["clean_text"])

def tokenize(example):
    example["sequence"] = tokenizer.texts_to_sequences([example["clean_text"]])[0]
    return example
dataset = dataset.map(tokenize)

print(dataset[0])
```

pad and truncate sequence

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences

max_length = 50
padded_sequences = pad_sequences(
    dataset["sequence"],
    maxlen=max_length,
    padding="post",
    truncating="post"
)

print("Shape:", padded_sequences.shape)
print(padded_sequences[:5])
```

Convert Labels to Categorical Vectors

```python
import numpy as np
from tensorflow.keras.utils import to_categorical
labels = np.array(dataset["label"])
num_classes = len(np.unique(labels))
print("Number of Classes:", num_classes)
one_hot_labels = to_categorical(labels, num_classes=num_classes)
print("Shape of One-Hot Labels:", one_hot_labels.shape)
for i in range(5):
    print(f"Original Label: {labels[i]}")
    print(f"One-Hot Vector: {one_hot_labels[i]}")
    print("-" * 40)
```

building the rnn model

```python

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, Dense

vocab_size = len(tokenizer.word_index) + 1
embedding_dim = 128
max_length = 50
num_classes = one_hot_labels.shape[1]

model = Sequential()

model.add(
    Embedding(
        input_dim=vocab_size,
        output_dim=embedding_dim,
        input_length=max_length
    )
)

model.add(
    SimpleRNN(
        units=64,
        activation="tanh"
    )
)
for each class
model.add(
    Dense(
        units=num_classes,
        activation="softmax"
    )
)

model.summary()
```

complinig the model

```python
# Import required library
from tensorflow.keras.optimizers import Adam
model.compile(
    optimizer=Adam(learning_rate=0.001),
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

history = model.fit(
    padded_sequences,
    one_hot_labels,
    epochs=10,
    batch_size=32,
    validation_split=0.2,
    shuffle=True,
    verbose=1
)


print("Training Completed Successfully!")
```

```python
loss, accuracy = model.evaluate(
    padded_sequences,
    one_hot_labels,
    verbose=1
)

print("\nModel Evaluation")
print("-" * 30)
print("Loss     :", loss)
print("Accuracy :", accuracy)


print("\nFinal Training Accuracy :", history.history["accuracy"][-1])
print("Final Validation Accuracy :", history.history["val_accuracy"][-1])

print("Final Training Loss :", history.history["loss"][-1])
print("Final Validation Loss :", history.history["val_loss"][-1])
```
